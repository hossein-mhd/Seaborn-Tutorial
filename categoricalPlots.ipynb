```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

## countplot

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.countplot(data=df, x='division')
```

```python
# with ylim you can choose the limitation of y axis range(the starting and ending range) 

plt.figure(figsize=(12, 4), dpi=200)
sns.countplot(data=df, x='division')

plt.ylim(90,260)
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.countplot(data=df, x='level of education')
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division')
```

## barplot

```python
# by default the estimator is 'mean'
# the 'ci' is use for the line on each bar. here it stands for standard deviation

plt.figure(figsize=(12, 4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd')
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')

plt.legend(bbox_to_anchor=(1.05,1))

plt.savefig('barplot.jpg')
```
