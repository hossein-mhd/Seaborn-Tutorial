```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

## boxplot()

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='math score', x='parental level of education')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='math score', x='parental level of education', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

## violinplot()

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='math score', x='parental level of education', hue='test preparation course',
              palette='Set2')
```

```python
# the 'split' split up the two violin
# the 'inner' delete the median and the bar in the center of the violin

plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df, y='math score', x='parental level of education', hue='test preparation course',
              palette='Set2', split=True, inner=None)
```

## swarmplot()

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y= 'gender', size=2, palette='Set2')
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y= 'gender', size=2, palette='Set2', hue='test preparation course')

plt.legend(bbox_to_anchor=(1.35,0.5))
```

```python
# the 'dodge' split the hue

plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y= 'gender', size=2, palette='Set2', hue='test preparation course',
             dodge=True)

plt.legend(bbox_to_anchor=(1.35,0.5))
```

## boxenplot()

```python
plt.figure(figsize=(8,4), dpi=200)
sns.boxenplot(x='math score', y='test preparation course', data=df, hue='gender')

plt.legend(bbox_to_anchor=(1.35,0.5))
```
