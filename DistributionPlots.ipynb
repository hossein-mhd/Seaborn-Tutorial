```python
sns.displot(x= 'salary', data= df)
```

```python
sns.set(style='darkgrid')
sns.displot(x= 'salary', data= df, bins=100)
```

## displot()

```python
# adding 'kde' and 'rug'

sns.set(style='darkgrid')
sns.displot(x= 'salary', data= df, bins=20, color='red', edgecolor='black', kde='True', rug='True')
```

## histplot()

the same result as displot()

```python
sns.set(style='darkgrid')
sns.displot(x= 'salary', data= df, bins=20, color='red', edgecolor='black', linewidth=2, ls= '--')
```

## kdeplot()

```python
sns.kdeplot(x= 'salary', data=df)
```
