```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('country_table.csv')
```

```python
df.shape
```

```python
df
```

```python
df = df.set_index('Countries')
```

```python
df
```

## heatmap()

```python
sns.heatmap(df)
```

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1))
```

```python
# 'linewidths' change the linewidths between cells.
# 'annot' write the values of the cells on it.
# 'cmap' is the same as 'paletter'.

plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, fmt='.1f', cmap='viridis')
```

## clustermap()

```python
sns.clustermap(df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis')
```

```python
sns.clustermap(df.drop('Life expectancy', axis=1), linewidths=0.5, annot=True, cmap='viridis', 
               col_cluster=False)
```
