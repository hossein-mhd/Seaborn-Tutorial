```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
# dpi sharpen up the image

plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
# hue, colorize the datapoints based on the column you give it to

plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='sales')
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='training level')
```

```python
# palette give us opportunity to choose what colors you want to use. you can choose the palette
# from matplotlib website in colormap section

plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='training level', palette='Dark2')
```

```python
# size, will make the points bigger base on the column you give

plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, size='salary')
```

```python
# the 's' stands for general size of the points.
# the 'alpha' use for controling transparency of the points. we can use to see the points which are
# under other points. the min is '0' and the max is '1'.

plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, s=200, alpha=0.2)
```

```python
# the style parameter uses the shapes to seprate the points base on the column you give.
# NOTICE : ONLY USE STYLE FOR CATEGORICAL COLUMNS

plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, s=200, style='level of education', hue='level of education')

plt.savefig('salesPlot.jpg')
```
